```python
import gradio as gr
from transformers import AutoModelForSequenceClassification, AutoTokenizer
import numpy as np
from scipy.special import softmax

# Load pre-trained model and tokenizer from Hugging Face
model_path = "cardiffnlp/twitter-roberta-base-sentiment-latest"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path)

def sentiment_analysis(text):
    # Tokenize input text
    encoded_input = tokenizer(text, return_tensors='pt')
    # Get model predictions
    output = model(**encoded_input)
    # Convert output to probabilities
    scores = output.logits.detach().numpy()
    scores = softmax(scores)
    # Return sentiment labels based on scores
    labels = ['negative', 'neutral', 'positive']
    return {labels[i]: float(scores[0][i]) for i in range(len(labels))}

# Create Gradio interface
iface = gr.Interface(fn=sentiment_analysis,
                     inputs="text",
                     outputs="label",
                     title="Sentiment Analysis App",
                     description="Enter text to analyze its sentiment.")

# Launch the app
iface.launch()
```

```python
import gradio as gr
from transformers import pipeline
```

```python
spam_detector = pipeline("text-classification", model="distilbert-base-uncased")
```

```python
def classify_spam(text):
    result = spam_detector(text)
    label = result[0]['label']
    score = result[0]['score']
    return f"{label} (Confidence: {score:.2f})"
```

```python
interface = gr.Interface(
    fn=classify_spam,
    inputs="text",
    outputs="text",
    title="Spam Detection",
    description="Enter a message to check if it's spam or not."
)

# Launch the interface
interface.launch()
```

```python
from transformers import pipeline

# Load POS tagging model specifically trained for POS tagging
pos_tagger = pipeline("ner", model="vblagoje/bert-english-uncased-finetuned-pos")

# Define POS tagging function
def tag_pos(input_text):
    results = pos_tagger(input_text)

    # Extracting words and their corresponding tags
    pos_tags = [(result['word'], result['entity']) for result in results]

    return pos_tags

# Test the function
if __name__ == "__main__":
    input_text = "The quick brown fox jumps over the lazy dog."
    pos_tags = tag_pos(input_text)

    # Print the results
    for word, tag in pos_tags:
        print(f"{word}: {tag}")
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Load model and tokenizer for grammar correction
model_name = "vennify/t5-base-grammar-correction"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# Define correction function
def correct_text(input_text):
    # Preprocess the input text
    input_ids = tokenizer.encode("grammar: " + input_text, return_tensors="pt")

    # Generate corrected text
    outputs = model.generate(input_ids)

    # Decode the generated output
    corrected_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return corrected_text

# Test the function
if __name__ == "__main__":
    input_text = "She go to the store yesterday."
    corrected_text = correct_text(input_text)

    # Print the results
    print("Original Text:", input_text)
    print("Corrected Text:", corrected_text)
```

```python
from transformers import pipeline
import gradio as gr

# Load the translation pipeline
translator = pipeline("translation_en_to_es", model="Helsinki-NLP/opus-mt-en-es")

# Define the translation function
def translate_text(text):
    result = translator(text)
    return result[0]['translation_text']

# Create the Gradio interface
demo = gr.Interface(
    fn=translate_text,
    inputs="text",
    outputs="text",
    title="Text Translation Demo",
    description="Translate English text to Spanish using Hugging Face's Transformers."
)

# Launch the application
demo.launch()
```
